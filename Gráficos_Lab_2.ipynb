##1) Procesamiento de datos (Subida)



```python
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
```

```python
dataset_1 = pd.read_excel("Serie 1 Subida Datos.xlsx")
```

```python
ax = sns.scatterplot(data=dataset_1, x= "Tiempo(s)", y = "Vector velocidad (m/s)", hue = "Vector velocidad (m/s)")
ax.set_title("Gráfico 1: Velocidad v/s tiempo(Subida)")
```

```python
tiempo = dataset_1.iloc[:7, 0]
velocidad = dataset_1.iloc[:7, 1]
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(tiempo, velocidad)
ax_0 = sns.regplot(x = tiempo, y = velocidad, color = "g", line_kws={'label':"y={0:.3f}x+{1:.3f}".format(slope,intercept)})
ax_0.set_title("Gráfico 2: Tramo de aceleración constante(Subida)")
ax_0.legend(loc="best")
```

```python
ax_1 = sns.scatterplot(data=dataset_1, x= "Tiempo(s)", y = "Fuerza 1 (N)", hue = "Fuerza 1 (N)")
ax_1.set_title("Gráfico 3: Fuerza v/s tiempo(Subida)")
```
